```python
import pandas as pd
import numpy as np

#np.random.seed()
```

np.random.seed(int) 

Genera numeri casuali partendo da un dato intero. In realtà i numeri generati sono pseudo casuali perchè partendo dallo stesso seed si ottiene la stessa serie di numeri.

Questo permette di testare algoritmi diversi con le stesse serie di num 'casuali'

```python
np.random.seed(666)
mydata = np.random.randint(0,101,(4,3)) # genera una matrice di valori casuali 4 righe 3 colonne
mydata
```

```python
myindex = ['CA','NY','AZ','TX'] #num righe = numero indici
```

```python
mycolumns = ['Jan','Feb','Mar'] #num colonne = numero labels colonne (elementi in mycolumns)
```

abbiamo quello che serve per creare un semplice dataframe usando i dati casuali della matrice mydata le colonne mycolumns e gli indice righe con myindex

### Per creare il dataframe

devo accedere alla libreria pandas e richiamare la classe DataFrame istanziando così un oggetto dataframe

```python
mydataframe = pd.DataFrame(data=mydata, index=myindex,columns=mycolumns) #Esplicitando il nome dei parametri sono sicuro di passargli le variabili corrette 
```

per il passaggio di parametri col nome vedi [keywords arguments su w3school](https://www.w3schools.com/python/gloss_python_function_arbitrary_keyword_arguments.asp)

```python
mydataframe #per convenzione dovrebbe chiamarsi df ma questo è il primo e volevo essere esplicitissimo
```

```python
mydataframe.info()
```

matplotlib è una libreria modulare quindi posso aggiungere i moduli man mano che ho bisogno di aggiugere informazioni da rappresentare.

- titolo
- griglia

Le librerie più evolute partono gia da grafici più completi, con matplotlib devo aggiungere ogni cosa


```python
plt.scatter(x,y)
plt.xlabel('ciao');
plt.ylabel('altra label');
```

```python
#Due grafici
plt.plot([2,4,6,7,9],[0,4,2,8,9], 'r--')
plt.plot([1,3,5,0,1],[0,4,2,8,9], 'g-')
```

```python
plt.axis([0,40,0,20]) #definisce gli estremi degli assi se non ci piacciono quelli automatici che calcola dai valori min e max
#        ^asseX ^
#               |asseY
```

```python
# Linspace ci permette di creare un dominio distribuito linearmente in un intervallo di valori

x = np.linspace(0, 10, 100) # asse x con 100 valori tra 0 e 10
y = np.exp(x) # calcolo i valori della fz esponenziale in corrispondenza di ogni punto sull'asse x
plt.plot(x,y)
plt.plot(-x,y)
```

#### Figure 
permette di cambiare le dimensione del plot ovvero della finestra dove compare il grafico

Va dichiarato **prima** del plot 

```python
plt.figure(figsize=(18,21))
x = np.linspace(0,100,1000)
y = np.sin(np.exp(x))
z = np.exp(x)
plt.plot(x,y)
plt.plot(x,z)
```

```python
# Per salvare i grafici in formato immagine c'è un metodo
plt.plot(x,y)
plt.axis([0,100,-5, 10])
plt.savefig('test.png')
plt.plot(x,z)
```
