```python
%matplotlib inline
from matplotlib import pyplot as plt
import numpy as np
import pandas as pd
```

- spine sono i bordi e le linee degli assi (spine) che stanno intorno o dentro al grafico hanno vari metodi tipo set_position e set_color
- xaxis e yaxis sono gli assi e anche loro hanno diversi metodi per il modo in cui visualizzare le 'tacche'

```python
x = np.linspace(-np.pi, np.pi, 500)
s = np.sin(x)
fig = plt.figure(figsize=(20,20))

r, c = 2, 3 

seno = fig.add_subplot(r, c,1)
seno.plot(x,np.abs(s), label='sin(|x|)')
seno.spines['bottom'].set_position('center')
seno.spines['left'].set_position('center')
seno.spines['top'].set_color(None)
seno.spines['right'].set_color(None)
seno.legend(loc='upper left')


coseno = fig.add_subplot(r, c,2)
coseno.plot(x, np.cos(x))
coseno.spines['bottom'].set_position('center')
coseno.spines['left'].set_position('center')
coseno.spines['top'].set_color(None)
coseno.spines['right'].set_color(None)

tg = fig.add_subplot(r, c, 3)
tg.plot(x, np.tan(x))
tg.spines['bottom'].set_position('center')
tg.spines['left'].set_position('center')
tg.spines['top'].set_color(None)
tg.spines['right'].set_color(None)

se = fig.add_subplot(r, c, 5)
se.plot(x, np.sin(np.exp(x)))
se.spines['bottom'].set_position('center')
se.spines['left'].set_position('center')
se.spines['top'].set_color(None)
se.spines['right'].set_color(None)



```

## le cose difficili alla fine

I File python possono essere lanciati come runtime oppure importati come librerie.

Se in uno stesso file voglio fare una libreria e poi anche farci delle operazioni da lanciare in runtime (ovvero avere chiamate di funzioni) devo inserire una condizione particolare.

La condizione da inserire è `if __name__ == "__main__:`

se il nome del file in esecuzione è main, allora il file è stato lanciato direttamente e vengono eseguite le chiamate di funzione.
Se il nome file non è main, il file .py sta girando come modulo dentro un altro file quindi non devo eseguire le chiamate di funzione ma solo permettere l'accesso alla libreria delle funzioni. Quindi tutto quello che sta dopo la riga if non verra eseguito. 

### Funzione Main

La funzione main() contiene tutta la logica di esecuzione del programma, le dichiarazioni di funzioni vanno prima del main e subito sotto la chiusura del blocco main ci sarà la condizione __name__ == '__main__' e la chiamata alla fz main()
