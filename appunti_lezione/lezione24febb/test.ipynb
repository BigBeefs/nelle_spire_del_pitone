### Files .py
In un file .py ci va solo il codice scritto in python.
Per eseguire un file di python, da terminale di sistema lancio il comando python 
seguito dal percorso relativo del file che vogliamo far girare.
esempio 
`C:\Users\esempio> python ./cartella_progetto/file_eseguibile.py`

questo comando lancia *file_eseguibile.py* che si trova in C:\Users\esempio\cartella_progetto\file_eseguibile.py 
e il suo relative path è ./cartella_progetto/file_eseguibile.py

**N.B.:** Si usano i path relativi perchè così quando sposto la cartella di lavoro
(per metterla in cloud o per condividere il lavoro con altri), il codice non ha bisogno di essere modificato andando a correggere tutti i path di ogni file.

# Variabile in python

- si usano per salvare in memoria i valori (dati) che ci servono per risolvere il problema -*occhio a quanta memoria usi*-
-Tipi di dati:
    - integer: numeri interi
    - float: numeri con virgola
    - string: caratteri alfanumerici
    - bool: True | False 

- dichiarazione variabile: in python la sintassi è nome valido = dato non è obbligatorio dichiarare il tipo.
Si può chiedere il tipo dato di una variabile con la funzione  `type(nome_variabile)`

```python
x = 42 
y = 3.14 
z = "hello World!" # STRINGHE SOLO CON VIRGOLETTE DOPPIE

print("la variabile x è di tipo", type(x))


print("la variabile y è di tipo", type(y))


print("la variabile z è di tipo", type(z))
```

### Dichiarare una variabile
- Regole del nome:
    - non può iniziare con un numero `~~1variablename = 1~~`

    - non si possono usare gli spazi al massimo il '_' 
    - mai usare '-' perchè è un operatore

- Conviene sempre usare nomi significativi
- meglio usare lowercase
- se la variabile ha 2 parole meglio usare underscore per leggibilità `la_variabile = 0`

*Oss: in notebook se chiamo la variabile la mostra a schermo (solo l'ultima chiamata) senza bisogno di print, se sto scrivendo un file .py devo invocare il print.*

```python
z 
```

```python
type(z)
```

Le variabili numeriche possono diventare stringhe con la funzione `str()`


```python
frase = "la mia età è: "
eta = 34
```

Concatenazione stringhe con l'opratore '+':

`frase + eta` 

genera errore perchè tipi file differenti

```python
frase + str(eta)
```

**N.B. la variabile eta rimane un int ma viene convertita solo nella riga sopra in cui l'ho passato alla funzione str()

```python
type(eta)
```

- caratteri speciali: le stringhe hanno caratteri che non sono visualizzabili a video ma cambiano il testo. Esempio `'\n'` va a capo
